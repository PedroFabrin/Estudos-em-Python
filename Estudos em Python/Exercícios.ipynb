```python
#1. Escreva um código que lê a lista abaixo e faça:
#A leitura do tamanho da lista
#A leitura do maior e menor valor
#A soma dos valores da lista

lista = [16, 14, 63, 65, 17, 99, 70, 11, 20, 48, 79, 32, 17, 89, 12, 25, 66]
```

```python
def leitura(lista):
  tamanho = len(lista)
  maior = max(lista)
  menor = min(lista)
  soma = sum(lista)
  return tamanho, maior, menor, soma
```

```python
tamanho, maior, menor, soma = leitura(lista)
```

```python
print(f"A lista possui {tamanho} números em que o maior número é {maior} e o menor número é {menor}. A soma dos valores presentes nela é igual a {soma}")
```

```python
#3. Crie a função que leia a lista abaixo e retorne uma nova lista com os
# múltiplos de 3:

lista = [97, 80, 94, 88, 80, 1, 16, 53, 62, 32, 24, 99]
```

```python
def funcao3(lista):
  nova_lista = []
  for num in lista:
    if num % 3 == 0:
      nova_lista.append(num)
  return nova_lista
```

```python
#5. Você foi contratado(a) como cientista de dados de uma associação de skate.
# Para analisar as notas recebidas de skatistas em algumas competições
# ao longo do ano, você precisa criar um código que
#calcula a pontuação dos(as) atletas. Para isso, o seu código deve receber
# 5 notas digitadas pelas pessoas juradas.


lista_notas = []

while len(lista_notas) < 5:
  nota = float(input("Digite a nota: "))
  lista_notas.append(nota)
```

```python
def calcularNota(lista):
  lista.remove(min(lista))
  lista.remove(max(lista))
  nota_final = sum(lista) / len(lista)
  return nota_final
```

```python
media = calcularNota(lista_notas.copy())
print(f'Nota da manobra: {media}')
```
